# Check Features Correlation

```python
# convert Level feature to numerical so that it can be correlated with others
patientsDf["Level"].replace({"Low":0,"Medium":1,"High":2}, inplace=True)
```

```python
#attributes = ["Age", "Gender", "Air Pollution", "Alcohol use", "Dust Allergy", "OccuPational Hazards"]
#corr_matrix = patientsDf.corr(attributes)

#extract all numerical features and check their correlation
numerical_features = patientsDf.select_dtypes(include=['number']).columns
corr_matrix = patientsDf[numerical_features].corr()
corr_matrix
```

**Heatmap of feature correlations**

```python
import seaborn as sns

plt.figure(figsize=(15,5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap of feature correlations')
plt.show()
```

Top Correlation values per "Level"

```python
corr_matrix["Level"].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix

#get names of top 5 correlated features
top_5_correlated_features = corr_matrix["Level"].sort_values(ascending=False).head(5).index

#draw scatter matrix of top 5 features
scatter_matrix(patientsDf[top_5_correlated_features], figsize=(12,8), alpha=.02)
```

```python
#let's focus on a specific plot
#patientsDf.plot(kind="scatter", x="Obesity", y="Alcohol use", alpha=0.005)
from collections import Counter

c = Counter(zip(patientsDf['Obesity'], patientsDf['Alcohol use']))

#Create a list of sizes based on occurrences, scaled for visibility
s = [2 * c[(xx, yy)] for xx, yy in zip(patientsDf['Obesity'], patientsDf['Alcohol use'])]

#Create the scatter plot with size based on occurrences
patientsDf.plot(kind="scatter", x="Obesity", y="Alcohol use", s=s, alpha=0.02)

plt.axis([-1, 10, -1, 10])
```

# Check Features Correlation After Scaling

```python
from sklearn.preprocessing import MinMaxScaler

minMaxScaler = MinMaxScaler()

patientsDf_num = patientsDf.select_dtypes(include=['number'])

X = patientsDf_num.drop("Level", axis=1)
y = patientsDf_num["Level"]
y = y.values.reshape(-1,1)

X_scaled = minMaxScaler.fit_transform(X)
y_scaled = minMaxScaler.fit_transform(y)

#create single dataframe from X_scaled and y_scaled
patientsDf_scaled = pd.DataFrame(X_scaled, columns=X.columns)
patientsDf_scaled["Level"] = y_scaled

corr_matrix_scaled = patientsDf_scaled.corr()
corr_matrix_scaled["Level"].sort_values(ascending=False)
```

note that the correlation values are the same even after scaling

```python
patientsDf_scaled.describe()
```

```python
from pandas.plotting import scatter_matrix

#get names of top 5 correlated features
top_5_correlated_features_scaled = corr_matrix_scaled["Level"].sort_values(ascending=False).head(5).index

#draw scatter matrix of top 5 features
scatter_matrix(patientsDf_scaled[top_5_correlated_features_scaled], figsize=(12,8), alpha=.02)
```

Also note that the scatter of the scaled dataframe is the same as non-Scaled

# Check Features Correlation Fer Each Class Label

```python
#create new dataset where patientsDf_scaled["Level"] convert 0 to 1 and all others to 1
patientsDf_scaled_low = patientsDf_scaled.copy()
patientsDf_scaled_low["Level"] = patientsDf_scaled_low["Level"].apply(lambda x: 1 if x == 0 else 0)

patientsDf_scaled_medium = patientsDf_scaled.copy()
patientsDf_scaled_medium["Level"] = patientsDf_scaled_medium["Level"].apply(lambda x: 1 if x == .5 else 0)

patientsDf_scaled_high = patientsDf_scaled.copy()
patientsDf_scaled_high["Level"] = patientsDf_scaled_high["Level"].apply(lambda x: 1 if x == 1 else 0)

corr_matrix_low_Class = patientsDf_scaled_low.corr()
corr_matrix_low_Class["Level"].sort_values(ascending=False)
```

```python
corr_matrix_medium_Class = patientsDf_scaled_medium.corr()
corr_matrix_medium_Class["Level"].sort_values(ascending=False)
```

```python
corr_matrix_high_Class = patientsDf_scaled_high.corr()
corr_matrix_high_Class["Level"].sort_values(ascending=False)
```

Notice how each class correlates differently with the features as we can see in the correaltion values above

# Create 3 Dataset of top 2,5,10 features of each class

We can see visually that the top10 correlated features for:
Low Level = []
Medium Level = []
High Level = []


```python
#create Create 3 Dataset of top 2,5,10 features of each class exluding Level and Index features

#exclude certain features
features_to_exclude = ["Level", "index"]

#sort correlation features in accodance with Level ignoring negative sign and excluding the level Itself and index as well as it's not relevant
top_2_correlated_features_low_Class = corr_matrix_low_Class["Level"].drop(features_to_exclude).apply(lambda x: abs(x)).sort_values(ascending=False).head(2).index
top_2_correlated_features_low_Class
```

```python
top_2_correlated_features_medium_Class = corr_matrix_medium_Class["Level"].drop(features_to_exclude).apply(lambda x: abs(x)).sort_values(ascending=False).head(2).index
top_2_correlated_features_medium_Class
```

```python

top_2_correlated_features_high_Class = corr_matrix_high_Class["Level"].drop(features_to_exclude).apply(lambda x: abs(x)).sort_values(ascending=False).head(2).index
top_2_correlated_features_high_Class
```

**Compose new dataset of top 6 correlated features (top 2 of each class)**

```python
#merge all the top2 features of the 3 classes and remove the duplicates
top_2_correlated_features_all_classes = top_2_correlated_features_low_Class.append(top_2_correlated_features_medium_Class).append(top_2_correlated_features_high_Class).drop_duplicates()

#create the dataset of the 6 features including the Level target feature
patientsDf_scaled_top_6 = patientsDf_scaled[top_2_correlated_features_all_classes.tolist() + ["Level"]]
patientsDf_scaled_top_6
```

**Compose new dataset of top 15 correlated features (top 5 of each class)**


```python
#exclude certain features
features_to_exclude = ["Level", "index"]

#sort correlation features in accodance with Level ignoring negative sign and excluding the level Itself and index as well as it's not relevant
top_5_correlated_features_low_Class = corr_matrix_low_Class["Level"].drop(features_to_exclude).apply(lambda x: abs(x)).sort_values(ascending=False).head(5).index

top_5_correlated_features_medium_Class = corr_matrix_medium_Class["Level"].drop(features_to_exclude).apply(lambda x: abs(x)).sort_values(ascending=False).head(5).index

top_5_correlated_features_high_Class = corr_matrix_high_Class["Level"].drop(features_to_exclude).apply(lambda x: abs(x)).sort_values(ascending=False).head(5).index

#merge all the top5 features of the 3 classes and remove the duplicates
top_5_correlated_features_all_classes = top_5_correlated_features_low_Class.append(top_5_correlated_features_medium_Class).append(top_5_correlated_features_high_Class).drop_duplicates()

#create the dataset of the 15 features including the Level target feature
patientsDf_scaled_top_15 = patientsDf_scaled[top_5_correlated_features_all_classes.tolist() + ["Level"]]
patientsDf_scaled_top_15
```

**Compose new dataset of top 30 correlated features (top 10 of each class)**

```python
#exclude certain features
features_to_exclude = ["Level", "index"]

#sort correlation features in accodance with Level ignoring negative sign and excluding the level Itself and index as well as it's not relevant
top_10_correlated_features_low_Class = corr_matrix_low_Class["Level"].drop(features_to_exclude).apply(lambda x: abs(x)).sort_values(ascending=False).head(10).index

top_10_correlated_features_medium_Class = corr_matrix_medium_Class["Level"].drop(features_to_exclude).apply(lambda x: abs(x)).sort_values(ascending=False).head(10).index

top_10_correlated_features_high_Class = corr_matrix_high_Class["Level"].drop(features_to_exclude).apply(lambda x: abs(x)).sort_values(ascending=False).head(10).index

#merge all the top5 features of the 3 classes and remove the duplicates
top_10_correlated_features_all_classes = top_10_correlated_features_low_Class.append(top_10_correlated_features_medium_Class).append(top_10_correlated_features_high_Class).drop_duplicates()

#create the dataset of the 30 features including the Level target feature
patientsDf_scaled_top_30 = patientsDf_scaled[top_10_correlated_features_all_classes.tolist() + ["Level"]]
patientsDf_scaled_top_30
```
